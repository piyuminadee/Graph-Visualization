```python
df = pd.read_csv("canada_per_capita_income.csv")
df
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.xlabel('Year')
plt.ylabel('Capital_income')
plt.scatter(df.year, df['per_capita_income_(US$)'], color = 'blue', marker = '+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']], df['per_capita_income_(US$)'])
```

```python
predicted_income_2024 = reg.predict([[2024]])
print(predicted_income_2024)
```
